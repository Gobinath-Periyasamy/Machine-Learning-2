### (2) i. Obtain the least squares solution wML to the regression problem using the analytical solution

```python
# import libraries

import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn import cluster,datasets, mixture
from sklearn.model_selection import train_test_split
from scipy.stats import multivariate_normal
from numpy import linalg as LA
```

### Train data

```python


# import Excel A2Q2 train data:

df_train = pd.read_csv("A2Q2Data_train.csv",header = None)
df_train
```

```python
df_train.ndim
```

```python
# Assign X_train, Y_train data
feature_train  = 0
X_train = df_train.iloc[:,:-1]
Y_train = df_train.iloc[::,-1:]
#iterating through the whole file
for row_train in X_train:
    feature_train= feature_train + 1
f_train = feature_train
n_train = len(df_train)
```

```python
print(f_train,n_train)
X_train = X_train.to_numpy()
y_train = Y_train.to_numpy()
```

```python
# Obtaining Analytical Solution (X^TX)^.X^Ty
w_ml = np.dot(np.dot(LA.inv(np.dot(X_train.T, X_train)),X_train.T),y_train)
```

```python
# Reshape W_ML
print("w_ml: ",w_ml.reshape(f_train),'\n')
```

```python
print("Mean Squared Error on train set using w_ML ",(1/n_train)*(LA.norm(X_train@w_ml - y_train))**2)
```

```python
# import Excel A2Q1 data:
df_test = pd.read_csv("A2Q2Data_test.csv",header = None)
df_test
```

### For Test set

```python
feature_test  = 0
X_test = df_test.iloc[:,:-1]
Y_test = df_test.iloc[::,-1:]
#iterating through the whole file
for row_test in X_test:
    feature_test = feature_test + 1
f_test = feature_test
n_test = len(df_test)
```

```python
print(f_test,n_test)
X_test = X_test.to_numpy()
y_test = Y_test.to_numpy()
```

```python
print("Mean Squared Error on test set using w_ML ",(1/n_test)*(LA.norm(X_test@w_ml - y_test))**2)
```
