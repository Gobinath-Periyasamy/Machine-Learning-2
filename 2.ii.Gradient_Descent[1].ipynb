### Gradient Descent Algorithm

```python
# import libraries

import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn import cluster,datasets, mixture
from sklearn.model_selection import train_test_split
from scipy.stats import multivariate_normal
from numpy import linalg as LA
```

### A2Q2 Train Data

```python
# import Excel A2Q2 data data:
df_train = pd.read_csv("A2Q2Data_train.csv",header = None)
df_train
```

### Gradient Descent Algorithm

```python
# Data import and Manipulation done in previous problem
feature_train  = 0
X_train = df_train.iloc[:,:-1]
Y_train = df_train.iloc[::,-1:]

f_train = X_train.shape[1]
n_train = len(df_train)
print(f_train,n_train)
X_train = X_train.to_numpy()
y_train = Y_train.to_numpy()
w_ml = np.dot(np.dot(LA.inv(np.dot(X_train.T, X_train)),X_train.T),y_train)
print("w_ml: ",w_ml.reshape(f_train),'\n')

print("Mean Squared Error on train set using w_ML ",(1/n_train)*(LA.norm(X_train@w_ml - y_train))**2)
```

```python
# import Excel A2Q1 data:
df_test = pd.read_csv("A2Q2Data_test.csv",header = None)
df_test
```

```python
# Assign X_train, Y_train data

feature_test  = 0
X_test = df_test.iloc[:,:-1]
Y_test = df_test.iloc[::,-1:]
#iterating through the whole file
for row_test in X_test:
    feature_test = feature_test + 1
f_test = feature_test
n_test = len(df_test)
```

```python
print(f_test,n_test)
X_test = X_test.to_numpy()
y_test = Y_test.to_numpy()
```

```python
#  least squares solution wML to compare with
print("Mean Squared Error on train set using w_ML ",(1/n_test)*(LA.norm(X_test@w_ml - y_test))**2)
```

### Gradient Descent Formula

```python
eta = 70e-7
w_grad = np.random.randn(f_train,1)
loss1 = []
diff = []

for i in range(1100):
    w_grad = w_grad -eta * (X_train.T@(X_train@w_grad - y_train))
    diff.append(LA.norm(w_grad - w_ml))
    
diff = np.array(diff)
diff  
```

### plot abs(w^t âˆ’ wML) Square Norm as a function of t

```python
plt.figure(figsize = (6,6))
plt.xlabel('time t')
plt.ylabel('Gradient Descent')
plt.title('Gradient Descent \n')
plt.plot(diff)
plt.grid()
# visualizing plot using matplotlib.pyplot library
plt.show()
```

```python
# Finding 
print("Final Mean Squared Error on train set using w_Grad ",(1/n_train)*(LA.norm(X_train@w_grad - y_train))**2)
```

```python
print("Final Mean Squared Error on test set using w_Grad ",(1/n_test)*(LA.norm(X_test@w_grad - y_test))**2)
```
